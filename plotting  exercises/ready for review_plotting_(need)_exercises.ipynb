### Exercise 1 

```python
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='blue', linestyle='dashed',
             marker='o', markerfacecolor='red',     # marker sets the shape of the marker, markerfacecolor is the colour of the market and ecolor is the error bar colour
             ecolor='black',capsize=5,
             label="y")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'upper left')
plt.show()
```

### Exercise 2 

```python
import numpy as np 
```

```python
# data 
x = np.linspace(0, 50, 5000)
y = np.sin(x) * np.exp(-x/10)

# plotting the graph 
plt.plot(x, y, c = 'blue')
plt.xlabel('x', fontsize = 15, c = 'darkorange')
plt.ylabel('y', fontsize = 15, c = 'darkorange', rotation = 'horizontal')    # add the rotation to reorientate the name of the axis, so it is facing the right way
plt.title(r'$y = \sin(x) \cdot e^{-x/10}$', fontsize = 20)         
plt.grid(alpha=0.25)
plt.ylim([-1, 1]) 

plt.show()
```
