## 1.3 Yet another plot but with error bars

````python
# data for plotting 
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, .75

# plotting 
plt.plot(x, y_1, label = 'y_1', 
        color = 'red', linestyle = 'dashed')
plt.plot(x, y_2, label = 'y_2',                
        color = 'blue')
plt.errorbar(x, y_3, xerr = x_error, yerr = y_error, 
            label = 'y_3 with errors', c = 'green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha = 0.25)
plt.legend(loc = 'upper left')
````

````python
# plotting data 
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# plotting
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, c = 'red', ls = 'dashed', label = 'y_1')
plt.plot(np_x, np_y_2, c = 'blue', label = 'y_2')
````

## 2.1 Adding mathematical functions

````python
# data
x = np.array([0, 1, 2, 3, 4, 5, 6])

# equations
x2 = x**2
sin_x = np.sin(x)

# plotting
plt.plot(x, x2, label = 'x^2', 
        c = 'red', ls = 'dashed')
plt.plot(x, sin_x, label = 'sin(x)', 
        c = 'blue')
plt.legend()         # auto is top left 
plt.show()
````

## We need another axis!

````python
# data and equations 
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

# first line and y axis
plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                               

# second line and corresponding y axis 
plt.twinx()                                                 
plt.plot(x, sin_x, label='sin(x)',color='blue', ) # for y-axis 2
````

### Things to note

````python
# data and equations 
x = np.linspace(0, 6, 100)     # 100 points between 0 and 6, for a smoother curve
x2 = x**2
sin_x = np.sin(x)

# first line and y axis
plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                               

# second line and corresponding y axis 
plt.twinx()                                                 
plt.plot(x, sin_x, label='sin(x)',color='blue', ) # for y-axis 2
````

# 3 Saving to disc

````python
# data points 
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

# plotting 
plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')

# second plot 
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

# storage 
plt.savefig('simple-plot.png', dpi=150)    # will save in the same file as this current jupyter notebook     # dpi = dots per inch 
````

# 4 A real example: Global Warming

## 4.1 Plotting data from files

````python
import numpy as np 
from matplotlib import pyplot as plt 
````

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
````

````python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
````

## 4.2 Styles

````python
# so you dont have to customise the figure 
plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
````
